```python
merged = r'merged.csv'
```

```python
df = pd.read_csv(merged)
print(df.columns)
# List of columns to plot
columns_to_plot = ['column1', 'column2', 'column3']

# Number of subplots
num_plots = len(columns_to_plot)

# Create subplots
fig, axes = plt.subplots(num_plots, 1, figsize=(10, 5 * num_plots))

# Loop through columns and create a subplot for each
for i, column in enumerate(columns_to_plot):
    axes[i].plot(df['truck_discharge_date'], df[column])
    axes[i].set_title(f'{column} vs truck_discharge_date')
    axes[i].set_xlabel('Truck Discharge Date')
    axes[i].set_ylabel(column)

# Adjust layout
plt.tight_layout()
plt.show()
```
